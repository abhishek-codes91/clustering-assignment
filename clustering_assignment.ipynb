Question 1: What is the difference between K-Means and Hierarchical Clustering?
Provide a use case for each.

Answer: Difference between K-Means and Hierarchical Clustering (in points)

1.Number of clusters

 - K-Means: Number of clusters (K) must be defined in advance

 - Hierarchical: No need to define clusters beforehand

2.Clustering method

 - K-Means: Centroid-based clustering

 - Hierarchical: Distance-based clustering (tree structure)

3.Process

 - K-Means: Iterative reassignment of points to nearest centroid

 - Hierarchical: Builds clusters step-by-step (bottom-up or top-down)

4.Output

 - K-Means: Flat set of clusters

 - Hierarchical: Dendrogram (hierarchy of clusters)

5.Scalability

 - K-Means: Fast and suitable for large datasets

 - Hierarchical: Slower, not suitable for very large datasets

6.Sensitivity

 - K-Means: Sensitive to initial centroids and outliers

 - Hierarchical: Sensitive to distance metric but not to initialization

 Question 2: Explain the purpose of the Silhouette Score in evaluating clustering
algorithms.

Answer:  Purpose of Silhouette Score (in points)

1.Measures cluster quality

 - Silhouette Score evaluates how well data points are clustered.

2.Checks cohesion

 - It measures how close a data point is to other points in the same cluster.

3.Checks separation

 - It measures how far a data point is from points in other clusters.

4.Score range

 - Value ranges from –1 to +1.

5.Interpretation

 - +1: Data point is well-clustered

- 0: Data point lies between two clusters

 - Negative value: Data point may be assigned to the wrong cluster

6.Model comparison

 - Helps compare different clustering algorithms or different values of K.

7.Optimal clusters

 - Used to find the best number of clusters by choosing the highest average silhouette score.

 Question 3: What are the core parameters of DBSCAN, and how do they influence the
clustering process?

Answer:  Core Parameters of DBSCAN and Their Influence (in points)

1.ε (Epsilon)

 - Defines the radius (neighborhood distance) around a data point.

 - Controls how close points must be to be considered neighbors.

 - Small ε: More clusters, many points marked as noise

- Large ε: Fewer clusters, may merge different clusters

2.MinPts (Minimum Points)

 - Minimum number of points required within ε to form a dense region.

 - Low MinPts: Sensitive to noise, may form false clusters

 - High MinPts: More strict clustering, more points labeled as noise

3.Distance Metric

 - Method used to calculate distance (e.g., Euclidean, Manhattan).

 - Affects how neighborhoods are formed and clusters are shaped.

 -How These Parameters Influence Clustering

 4.Cluster formation

 - Points with ≥ MinPts neighbors within ε become core points.

5.Border points

 - Points within ε of a core point but with fewer than MinPts neighbors.

6.Noise points

 - Points that do not satisfy either condition are labeled as outliers.

 Question 4: Why is feature scaling important when applying clustering algorithms like
K-Means and DBSCAN?

Answer:
-Why Feature Scaling is Important in Clustering (in points)

1.Distance-based algorithms

 - K-Means and DBSCAN use distance calculations to form clusters.

2.Prevents dominance of large-scale features

 - Features with larger numerical ranges can dominate distance calculations if not scaled.

3.Ensures equal contribution

 - Scaling ensures all features contribute equally to the clustering process.

4.Improves cluster accuracy

 - Leads to more meaningful and correctly formed clusters.

5.Better centroid calculation (K-Means)

 - Scaled features help compute accurate centroids.

6.Correct neighborhood detection (DBSCAN)

 - Scaling helps DBSCAN correctly identify ε-neighborhoods.

7.Faster and stable convergence

 - Especially in K-Means, scaling can lead to faster convergence.

 Question 5: What is the Elbow Method in K-Means clustering and how does it help
determine the optimal number of clusters?

Answer: Elbow Method is a technique used in K-Means clustering to determine the optimal number of clusters (K) by plotting the within-cluster sum of squares (WCSS) against different values of K and selecting the point where the decrease in WCSS starts to slow down, forming an “elbow” shape.

-How the Elbow Method Helps Determine the Optimal Number of Clusters (in points)

1.Calculates WCSS for different values of K.

2.Plots K vs WCSS on a graph.

3.Observes how WCSS decreases as K increases.

4.Identifies the point where the decrease in WCSS becomes less steep.

5.This point forms an “elbow” shape in the graph.

6.The K value at the elbow is chosen as the optimal number of clusters because adding more clusters gives little improvement.


Question 6: Generate synthetic data using make_blobs(n_samples=300, centers=4),
apply KMeans clustering, and visualize the results with cluster centers.

answer-

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic dataset
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

# Get cluster centers
centers = kmeans.cluster_centers_

# Visualize clusters and centers
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(centers[:, 0], centers[:, 1], marker='X', s=200)
plt.title("KMeans Clustering on Synthetic Data (make_blobs)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

#output
```

Question 7: Load the Wine dataset, apply StandardScaler , and then train a DBSCAN
model. Print the number of clusters found (excluding noise).

answer-

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Load Wine dataset
wine = load_wine()
X = wine.data

# Apply StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train DBSCAN model
dbscan = DBSCAN(eps=1.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Number of clusters (excluding noise)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

print("Number of clusters found (excluding noise):", n_clusters)

# output

Number of clusters found (excluding noise): 0
```

Question 8: Generate moon-shaped synthetic data using
make_moons(n_samples=200, noise=0.1), apply DBSCAN, and highlight the outliers in
the plot.
(Include your Python code and output in the code box below.)

Answer:

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

# Generate moon-shaped synthetic data
X, y = make_moons(n_samples=200, noise=0.1, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X)

# Identify outliers (noise points)
outliers = labels == -1
clusters = labels != -1

# Plot clusters and outliers
plt.scatter(X[clusters, 0], X[clusters, 1])
plt.scatter(X[outliers, 0], X[outliers, 1], marker='x')
plt.title("DBSCAN on make_moons Dataset (Outliers Highlighted)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Question 9: Load the Wine dataset, reduce it to 2D using PCA, then apply
Agglomerative Clustering and visualize the result in 2D with a scatter plot.
(Include your Python code and output in the code box below.)

Answer:

```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import numpy as np

# Load Wine dataset
wine = load_wine()
X = wine.data

# Reduce data to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=3)
labels = agg.fit_predict(X_pca)

# Visualize clusters
for cluster_id in np.unique(labels):
    plt.scatter(
        X_pca[labels == cluster_id, 0],
        X_pca[labels == cluster_id, 1]
    )

plt.title("Agglomerative Clustering on Wine Dataset (PCA Reduced)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

Question 10: You are working as a data analyst at an e-commerce company. The
marketing team wants to segment customers based on their purchasing behavior to run
targeted promotions. The dataset contains customer demographics and their product
purchase history across categories.
Describe your real-world data science workflow using clustering:
● Which clustering algorithm(s) would you use and why?
● How would you preprocess the data (missing values, scaling)?
● How would you determine the number of clusters?
● How would the marketing team benefit from your clustering analysis?
(Include your Python code and output in the code box below.)

answer-

-Real-World Clustering Workflow for Customer Segmentation
 1.Clustering Algorithm Choice

-K-Means Clustering

 - Efficient and scalable for large customer datasets

 - Easy to interpret and explain to non-technical teams

-(Optional) DBSCAN

 - Useful for detecting outlier customers (very high/low spenders)

 - Primary choice: K-Means (business-friendly and fast)

2.Data Preprocessing

-Missing values

 - Numerical: fill with mean/median

 - Categorical: fill with mode

-Encoding

 - Convert categorical variables (gender, city) using one-hot encoding

-Feature scaling

 - Apply StandardScaler

 - Required because clustering is distance-based

3.Determining Number of Clusters

-Elbow Method

 - Finds point where adding more clusters gives diminishing returns

-Silhouette Score

- Chooses the number of clusters with best separation and cohesion

4.Business Benefit for Marketing Team

- Identify high-value customers for premium offers

 - Target price-sensitive customers with discounts

 - Design personalized campaigns by segment

 - Improve customer retention and conversion rates

 - Optimize marketing budget allocation

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Simulated customer dataset
np.random.seed(42)
data = {
    "age": np.random.randint(18, 65, 200),
    "annual_income": np.random.randint(20000, 120000, 200),
    "electronics_spend": np.random.randint(0, 5000, 200),
    "fashion_spend": np.random.randint(0, 4000, 200),
    "grocery_spend": np.random.randint(0, 3000, 200)
}

df = pd.DataFrame(data)

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Find optimal clusters using Silhouette Score
scores = {}
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    scores[k] = silhouette_score(X_scaled, labels)

# Train final model
optimal_k = max(scores, key=scores.get)
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
df["Cluster"] = kmeans_final.fit_predict(X_scaled)

optimal_k, scores

#output
Optimal number of clusters: 5
Silhouette Scores: {
2: 0.154,
3: 0.152,
4: 0.143,
5: 0.161
}
```
